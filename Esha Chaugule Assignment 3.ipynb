# QUESTION 10

Grouping all depth overs

```python
depthoveranalysis=deliveries.loc[deliveries.over>14]
depthoveranalysis
```

```python
depthoveranalysis.groupby("over").apply(lambda a: a[:])
```

Total runs in the last 15 overs

```python
depthoveranalysis.groupby("over")["total_runs"].sum()
```

Total wickets in the last 15 overs

```python
depthoveranalysis.groupby("over")["is_wicket"].sum()
```

```python
depthoveranalysis.groupby(["inning", "over"]).agg({"batsman_runs":"sum", "is_wicket":"sum"})
```

Thus, we can conclude that runs towards the last overs in the first innings are greater, and so the batting team tries to increase the target as much as possible.
But in the second innings it's less because either the target is already reached and game is over, or all batsmen are out.
In the first innings, the bowling team tries to get as many wickets as possible, but in the second innings the wickets are less because the target has been reached or most batsmen have been out in the previous overs. 
This tells us that during the last overs of the first innings, the batsmen try to hit boundries, thus making more runs, but also in the process getting out.
